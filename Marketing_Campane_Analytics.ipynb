Starting WIth Data Cleaning 

```python
df['Income'] = df['Income'].fillna(
    df['Income'].median()
)
```

```python
df.duplicated().sum()
```

3.Converting Date Column

```python
df['Dt_Customer'] = pd.to_datetime(
    df['Dt_Customer']
)
```

```python
df.dtypes
```

Feature Engineering 

```python
df['Age'] = 2026 - df['Year_Birth']
```

```python
df['Children'] = (
    df['Kidhome']
    + df['Teenhome']
)
```

```python
df['Total_Spending'] = (
      df['MntWines']
    + df['MntFruits']
    + df['MntMeatProducts']
    + df['MntFishProducts']
    + df['MntSweetProducts']
    + df['MntGoldProds']
)
```

```python
df['Customer_Tenure'] = (
    2026 -
    df['Dt_Customer'].dt.year
)
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18,30,45,60,100],
    labels=[
        'Young Adult',
        'Adult',
        'Middle Age',
        'Senior'
    ]
)
```

```python
df['Income_Group'] = pd.qcut(
    df['Income'],
    q=3,
    labels=[
        'Low Income',
        'Middle Income',
        'High Income'
    ]
)
```

```python
df['Total_Purchases'] = (
      df['NumWebPurchases']
    + df['NumCatalogPurchases']
    + df['NumStorePurchases']
)
```

```python
df['Accepted_Campaigns'] = (
      df['AcceptedCmp1']
    + df['AcceptedCmp2']
    + df['AcceptedCmp3']
    + df['AcceptedCmp4']
    + df['AcceptedCmp5']
)
```

```python
df['Customer_Segment'] = np.where(
    df['Total_Spending'] >=
    df['Total_Spending'].median(),
    'High Value',
    'Low Value'
)
```
